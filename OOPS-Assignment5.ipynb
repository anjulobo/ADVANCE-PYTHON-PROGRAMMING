```python
#Decorators
#1 - Create a simple decorator that adds extra print statements before and after a function execution

def my_decorator(func):

    def wrapper():
        print("Before function execution")

        func()

        print("After function execution")

    return wrapper

@my_decorator
def greet():
    print("Hello, World!")

greet()
```

```python
#2 - Create a decorator that wraps a function and prints a custom message before calling the original function

def custom_message_decorator(func):

    def wrapper():
        print("Starting the function execution...")
        func()

    return wrapper

@custom_message_decorator
def display_message():
    print("Welcome to Python Decorators!")

display_message()
```

```python
#3 - Create a decorator that takes a function with two arguments and logs the function name along with its parameters

def log_function_call(func):

    def wrapper(a, b):
        print(f"Calling function: {func.__name__}")
        print(f"Parameters: a = {a}, b = {b}")

        return func(a, b)

    return wrapper

@log_function_call
def add(x, y):
    return x + y

result = add(10, 20)

print("Result:", result)
```

```python
#4 - Create a decorator to measure execution time of a function that takes no arguments.

import time

def measure_time(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()

        execution_time = end_time - start_time
        print(f"Execution Time: {execution_time:.6f} seconds")

    return wrapper

@measure_time
def sample_task():
    time.sleep(2)
    print("Task completed")

sample_task()
```

```python
#5 - Create a decorator that checks if the input user is "admin" before allowing function execution

def admin_required(func):
    def wrapper(username):
        if username == "admin":
            return func(username)
        else:
            print("Access Denied! this is not admin.")
    return wrapper

@admin_required
def delete_records(username):
    print(f"Welcome {username}! Records have been deleted.")

delete_records("admin")
delete_records("john")
```

```python
#6 - Create a decorator that caches results of a function with one integer argument to avoid repeated calculations

def cache_results(func):
    cache = {}

    def wrapper(n):
        if n in cache:
            print(f"Returning cached result for {n}")
            return cache[n]

        result = func(n)
        cache[n] = result
        print(f"Calculated and cached result for {n}")
        return result

    return wrapper

@cache_results
def square(n):
    return n * n

print(square(5))
print(square(5))
print(square(10))
print(square(10))
```

```python
#7 - Create a decorator that works with functions having arguments using *args and **kwargs and prints messages before and after execution

def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before function execution")

        result = func(*args, **kwargs)

        print("After function execution")
        return result

    return wrapper

@my_decorator
def add(a, b):
  print(f"Sum: {a + b}")
  return a + b

@my_decorator
def greet(name, message="Hello"):
    print(f"{message}, {name}!")

add(10, 20)
print()

greet(name="Anju", message="Welcome")
```

```python
#8 - Create a decorator that modifies the return value of a function by doubling the result

def double_result(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result * 2
    return wrapper

@double_result
def add(a, b):
    return a + b

@double_result
def multiply(a, b):
    return a * b

print(add(5, 3))
print(multiply(4, 2))
```
