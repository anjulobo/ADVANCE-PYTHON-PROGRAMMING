```python
#1 - Create an ElectricCar class that inherits from the Car class and includes an additional attribute, battery_size.

class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")

class ElectricCar(Car):
    def __init__(self, make, model, battery_size):
        super().__init__(make, model)
        self.battery_size = battery_size

    def display_info(self):
        super().display_info()
        print(f"Battery Size: {self.battery_size} kWh")

ev = ElectricCar("Tesla", "Model 3", 75)
ev.display_info()
```

```python
from types import ModuleType
#2 - Define a Laptop class that inherits from the Computer class and includes an additional attribute, battery_life.

class Computer:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")

class Laptop(Computer):
    def __init__(self, brand, model, battery_life):
        super().__init__(brand, model)
        self.battery_life = battery_life

    def display_info(self):
        super().display_info()
        print(f"Battery Life: {self.battery_life} hours")

laptop = Laptop("Dell", "Lenovo 15", 10)
laptop.display_info()
```

```python
from sys import modules
#3 - Create a Smartphone class and add another child class of your choice that inherits from an ElectronicDevice class.
# Include an additional attribute, operating_system.

class ElectronicDevice:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")

class Smartphone(ElectronicDevice):
    def __init__(self, brand, model, operating_system):
        super().__init__(brand, model)
        self.operating_system = operating_system

    def display_info(self):
        super().display_info()
        print(f"Operating System: {self.operating_system}")

class Tablet(ElectronicDevice):
    def __init__(self, brand, model, screen_size):
        super().__init__(brand, model)
        self.screen_size = screen_size

    def display_info(self):
        super().display_info()
        print(f"Screen Size: {self.screen_size} inches")

smartphone = Smartphone("Samsung", "Galaxy S26", "Android")
smartphone.display_info()

print("\n")

tablet = Tablet("Apple", "iPad Pro", 12.9)
tablet.display_info()
```

```python
#4 - Implement a Musician class and create another child class of your choice that extends an Artist class.
#Include an additional attribute, instrument_played.


class Musician:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"I am {self.name}, an artist.")


class Artist(Musician):
    def __init__(self, name, instrument):
        super().__init__(name)
        self.instrument = instrument

    def play_music(self):
        print(f"{self.name} plays the {self.instrument}.")

# Moved Painter class definition outside Artist.play_music method
class Painter(Artist):
    def __init__(self, name, instrument, style):
        super().__init__(name, instrument) # Artist takes name and instrument
        self.style = style

    def paint_art(self):
        print(f"{self.name} paints in the {self.style} style.")

# Test cases - moved outside methods and corrected instantiation
musician_obj = Musician("Arijit") # Corrected: Musician only takes 'name'
musician_obj.introduce()

artist_obj = Artist("Adele", "Vocals") # Create an Artist object
artist_obj.introduce()
artist_obj.play_music()

painter_obj = Painter("Vincent van Gogh", "Brush", "Post-Impressionism")
painter_obj.introduce()
painter_obj.play_music() # Inherited method
painter_obj.paint_art()
```

```python
#5 - Define a FictionBook class and add another child class of your choice that inherits from the Book class.
#Include an additional attribute, genre

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}")

class FictionBook(Book):
    def __init__(self, title, author, genre):
        super().__init__(title, author)
        self.genre = genre

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}, Genre: {self.genre}")

class ScienceBook(Book):
    def __init__(self, title, author, field):
        super().__init__(title, author)
        self.field = field

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}, Field: {self.field}")

f1 = FictionBook("The Hobbit", "J.R.R. Tolkien", "Fantasy")
s1 = ScienceBook("A Brief History of Time", "Stephen Hawking", "Physics")

f1.display_info()
s1.display_info()
```
