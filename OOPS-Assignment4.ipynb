```python
#Abstraction
#1 - Create an abstract class with one abstract method and two subclasses that implement this method with different outputs

from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def make_sound(self):
        pass

class Cow(Animal):

    def make_sound(self):
        print("Cow says: Moo-oo!")

class Cat(Animal):

   def make_sound(self):
        print("Cat says: Meow Meow!")

cow = Cow()
cat = Cat()

cow.make_sound()
cat.make_sound()
```

```python
#2 - Create an abstract class with one abstract method and one normal method, then create a subclass
# that implements the abstract method and calls both methods

from abc import ABC, abstractmethod

class Employee(ABC):

    @abstractmethod
    def work(self):
        pass

    def company_info(self):
        print("Welcome to AAFT University")

class Developer(Employee):

    def work(self):
        print("Creating intelligent minds.")

dev = Developer()

dev.company_info()

dev.work()
```

```python
#3 - Create an abstract class for vehicles with an abstract method for maximum speed and implement it in at least two subclasses

from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def maximum_speed(self):
        pass

class Car(Vehicle):

    def maximum_speed(self):
        print("Car maximum speed: 220 km/h")

class Bike(Vehicle):

  def maximum_speed(self):
        print("Bike maximum speed: 180 km/h")

car = Car()
bike = Bike()

car.maximum_speed()
bike.maximum_speed()
```

```python
#4 - Create an abstract class for shapes with an abstract method to calculate area and implement it in a subclass like circle

from abc import ABC, abstractmethod
import math

class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass

class Circle(Shape):

  def __init__(self, radius):
        self.radius = radius

  def calculate_area(self):
        area = math.pi * self.radius ** 2
        print("Area of Circle:", area)

circle = Circle(5)

circle.calculate_area()
```

```python
#5 - Create an abstract class with a method for security and another normal method for database connection,
# then implement the abstract method in a subclass and demonstrate all methods

from abc import ABC, abstractmethod

class SystemSecurity(ABC):

    @abstractmethod
    def authenticate_user(self):
        pass

    def connect_database(self):
        print("Database connected successfully.")

class LoginSystem(SystemSecurity):

    def authenticate_user(self):
        print("User authenticated using username and password.")

system = LoginSystem()

system.connect_database()

system.authenticate_user()
```
