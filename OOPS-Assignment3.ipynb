```python
#1 - Polymorphism (Method Based)
# Create two different classes that contain a method with the same name. Write a common function that accepts an object and calls
#that method to demonstrate polymorphism. Write another example where multiple objects are stored in a list and the same method
#is called using a loop to show how Python handles method calls dynamically.

# Class 1
class Dog:
    def speak(self):
        print("Dog says: Woof!")

# Class 2
class Goat:
    def speak(self):
        print("Goat says: Mee-e!")

# Common function
def make_sound(animal):
    animal.speak()

# Creating objects
dog = Dog()
goat = Goat()

# Calling the same function with different objects
make_sound(dog)
make_sound(goat)
```

```python
#2 - Polymorphism using Inheritance (Method Overriding)
 #Create a base class with a method. Then create one or more derived classes that override this method and show how different
 #objects call their own version of the method. Also demonstrate this concept using multiple child classes stored in a list
 #and executed using a loop.

class Book:
    def display_info(self):
        print("This is a generic book.")

class FictionBook(Book):
    def display_info(self):
        print("This is a Fiction Book with imaginative storytelling.")

class ScienceBook(Book):
    def display_info(self):
        print("This is a Science Book with factual information.")

class HistoryBook(Book):
    def display_info(self):
        print("This is a History Book with past events and analysis.")

b1 = Book()
f1 = FictionBook()
s1 = ScienceBook()
h1 = HistoryBook()

b1.display_info()
f1.display_info()
s1.display_info()
h1.display_info()

books = [FictionBook(), ScienceBook(), HistoryBook()]

for book in books:
    book.display_info()
```

```python
#3 - Operator Overloading
#Create a class that represents a 3D object and includes a method to calculate its volume. Overload the + operator
 # so that when two objects are added, the result gives the combined volume.

class ThreeDObject:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height


    def volume(self):
        return self.length * self.width * self.height

    def __add__(self, other):
        if isinstance(other, ThreeDObject):
            return self.volume() + other.volume()
        return NotImplemented

box1 = ThreeDObject(2, 3, 4)
box2 = ThreeDObject(5, 6, 2)

print("Box1 Volume:", box1.volume())
print("Box2 Volume:", box2.volume())

combined_volume = box1 + box2
print("Combined Volume:", combined_volume)
```

```python
#4 - Method Variation (Same Name, Different Parameters)
 #Create two classes where both have a method with the same name but accept different parameters. Demonstrate how the method
 #behaves differently based on the object and arguments passed.

class Rectangle:
    def area(self, length, width):
        return length * width

class Circle:
    def area(self, radius):
        return 3.14159 * radius * radius

rect = Rectangle()
circle = Circle()

print("Rectangle Area:", rect.area(5, 3))
print("Circle Area:", circle.area(4))
```
